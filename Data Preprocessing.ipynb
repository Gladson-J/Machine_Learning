### Normalization

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
data1=pds.DataFrame({'x':np.random.randint(-100,100,200),'y':np.random.randint(-50,50,200),'z':np.random.randint(-700,700,200)})
```

```python
pl=plt.axes(projection='3d')
```

```python
pl=plt.axes(projection='3d')
pl.scatter3D(data1.x,data1.y,data1.z)
plt.figure()
```

```python
from sklearn.preprocessing import Normalizer
```

```python
n=Normalizer()
```

```python
data2=n.fit_transform(data1)
```

```python
data2
```

```python
DF=pds.DataFrame(data2,columns=['x1','y1','z1'])
DF
```

```python
pl=plt.axes(projection='3d')
pl.scatter3D(DF.x1,DF.y1,DF.z1)
plt.figure()
```
