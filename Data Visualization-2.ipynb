```python
sb.get_dataset_names()
```

```python
tips=sb.load_dataset('tips')
```

```python
tips
```

## Relationship Plot 

```python
sb.relplot(x='total_bill', y='tip', data=tips, col='sex', hue='day',style='smoker', kind='scatter') 
```

```python
glue=sb.load_dataset('glue')
```

```python
glue
```

```python
sb.relplot(x='Task', y='Model', data=glue, col='Year', hue='Score',style='Encoder', kind='scatter')
```

### Categorical Plot

```python
sb.catplot(x='smoker', y='tip', data=tips, kind='box')
```

```python
sb.catplot(x='Score', y='Encoder', data=glue,kind='box')
```
